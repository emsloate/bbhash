```python
import random
```

```python
def create_kmer():
    kmer = ""
    letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(10):
        kmer += random.choice(letters)
    kmer += "\n"
    return kmer
```

```python
test_lengths = [10000,100000,1000000]
```

```python
for length in test_lengths:
    kmers = []
    for i in range(test_lengths):
        #pretty much guaranteed to have no duplicates
        kmers.append(create_kmer())

    f = open("./test_files/{}test.txt".format(length,"w")
    for line in kmers:
        f.write(line)
    f.close()
    
```

```python
#creating alien keys
for length in test_lengths:
    f = open("./test_files/{}test.txt".format(length),"r")
    kmers = f.readlines()
    
    num_alien = 0
    aliens = []
    for i in range(1000000):
        km = create_kmer()
        if num_alien == 1000:
            print("finished")
            break
        if km not in kmers:
            aliens.append(km)
            num_alien += 1
    f.close()
    
    #write to file
    f = open("/test_files/{}alien.txt".format(length),"w")
    for line in aliens:
        f.write(line)
    f.close()
```

## Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
speed_results = pd.read_csv("./test_files/speed_results.txt",names= ["gamma","n","time (ms)"])
#seaborn is stupid! treats gammas as continuous for hue, this is the only way to prevent that :/
speed_results["gamma"] = ["$%s$" % x for x in speed_results["gamma"]]
speed_results
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.lineplot(x='n',y='time (ms)',hue = 'gamma',data=speed_results, ax = ax)
ax.set_title("Average lookup time for values of n, gamma")
ax.set_xscale('log')
plt.show()
```

```python
size_results = pd.read_csv("./test_files/size_results.txt",names= ["gamma","n","bits","hash_table_size"])
#seaborn is stupid! treats gammas as continuous for hue, this is the only way to prevent that :/
size_results["gamma"] = ["$%s$" % x for x in size_results["gamma"]]
size_results
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.lineplot(x='n',y='bits',hue = 'gamma',data=size_results, ax = ax)
ax.set_title("Total number of bits for values of n, gamma")
ax.set_xscale('log')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.lineplot(x='n',y='hash_table_size',hue = 'gamma',data=size_results, ax = ax)
ax.set_title("Number of elements in hash table for values of n, gamma")
ax.set_xscale('log')
ax.set_yscale('log')
plt.show()
```

```python
alien_results = pd.read_csv("./test_files/alien_results.txt",names =['gamma','n','percent_detected'])
alien_results["gamma"] = ["$%s$" % x for x in alien_results["gamma"]]
alien_results['percent_detected'] = alien_results['percent_detected'] * 100
```
